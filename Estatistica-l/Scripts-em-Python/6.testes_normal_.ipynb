```python
# Importação das bibliotecas: matplotlib para geração de gráficos - Este código importa algumas bibliotecas em Python, incluindo scipy.stats, scipy.stats.norm, scipy.stats.skewnorm e matplotlib.pyplot. A biblioteca "scipy.stats" é usada para realizar cálculos estatísticos e criar distribuições estatísticas. As importações "norm" e "skewnorm" são para distribuições de probabilidade normais e assimétricas. A biblioteca "matplotlib.pyplot" é usada para criar gráficos e visualizações dos dados estatísticos. No código em si, podem seguir operações adicionais para análise e visualização de dados estatísticos usando as funções dessas bibliotecas.

from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
# Criação de uma variável com dados em uma distribuição normal com a função rvs (100 elementos) - O código gera uma amostra de 1000 números aleatórios seguindo uma distribuição normal (gaussiana) usando a função "norm.rvs" e armazena esses números na variável "dados". Em seguida, exibe a amostra de números gerados.

dados = norm.rvs(size = 1000)
dados
```

```python
# Histograma - Este código utiliza a biblioteca Matplotlib para criar um histograma dos dados contidos na variável "dados". Os dados são divididos em 20 intervalos (ou bins) para visualização. O comando "plt.title" define o título do gráfico como "Dados". Em resumo, o código gera um gráfico de histograma para representar a distribuição dos dados.

plt.hist(dados, bins = 20)
plt.title('Dados')
```

- A imagem acima é um gráfico de probabilidade, um tipo de análise gráfica usada para avaliar a distribuição de um conjunto de dados. 
- O gráfico é um gráfico de dispersão com uma linha azul conectando os pontos. 
- O eixo x está rotulado como “Quantis teóricos” e o eixo y está rotulado como “Valores ordenados”. 
- O gráfico parece ser linear, indicando que os dados são normalmente distribuídos. 
- Há alguns outliers no canto superior direito e inferior esquerdo do gráfico.

```python
# Geração de gráfico para verificar se a distribuição é normal - Esse código em Python cria um gráfico de probabilidade (QQ plot) usando a biblioteca Matplotlib e SciPy. Ele plota os dados do array 'dados' em relação a uma distribuição teórica (normal, por padrão) para verificar se os dados seguem essa distribuição. O gráfico é exibido na tela usando plt.show().

fig, ax = plt.subplots()
stats.probplot(dados, fit=True,   plot=ax)
plt.show()
```

- A imagem abaixo é um gráfico de probabilidade, que é uma técnica gráfica para avaliar se um conjunto de dados segue ou não uma determinada distribuição, como a distribuição normal. 
- O gráfico tem uma linha vermelha e pontos azuis. 
- O eixo x está rotulado como “Quantis teóricos” e o eixo y está rotulado como “Valores ordenados”. 
- A linha vermelha é uma linha reta, o que sugere que os dados seguem uma distribuição normal. 
- Os pontos azuis estão espalhados em torno da linha vermelha, o que sugere que há algum desvio da distribuição normal.

```python
# Execução do teste de Shapiro segundo argumento é o valor de p, não há como rejeitar a hipótese nula - Esse código em Python utiliza a função "shapiro" do módulo "stats" para realizar o teste de normalidade de uma variável chamada "dados". O teste de Shapiro-Wilk verifica se a distribuição dessa variável é normal. O resultado do teste ajudará a determinar se os dados seguem uma distribuição normal ou não.

stats.shapiro(dados)
```

Dados não normais

```python
# O código gera uma amostra de 1000 números aleatórios seguindo uma distribuição skewnorm com um parâmetro de assimetria de 4. Isso cria uma distribuição de dados que não é simétrica e tende a ser inclinada para a direita. A função "skewnorm.rvs" é usada para gerar esses valores aleatórios com base na distribuição especificada.

dados2 = skewnorm.rvs(4, size=1000)
```

```python
# Histograma - Esse código utiliza a biblioteca Matplotlib para criar um histograma dos dados contidos na variável 'dados2' com 20 intervalos (bins). O gráfico é então intitulado como 'Dados'.

plt.hist(dados2, bins = 20)
plt.title('Dados')
```

- A imagem acima é um gráfico de dados com um histograma azul e um fundo branco. 
- O histograma é uma representação gráfica da distribuição de frequências de um conjunto de dados. 
- Neste caso, o eixo x está rotulado com números de -0,5 a 3,5 e o eixo y está rotulado com números de 0 a 120. 
- O histograma tem um pico em torno de 1,5 no eixo x e em torno de 100 no eixo y, o que indica que a maioria dos dados está concentrada nessa região. 
- O histograma também tem uma cauda longa no lado direito, o que indica que há alguns valores mais altos no conjunto de dados.

```python
# Geração de gráfico para verificar se a distribuição é normal - Este código cria um gráfico de probabilidade (Q-Q plot) usando a biblioteca Matplotlib no Python. Ele exibe a distribuição dos dados contidos em 'dados2' em relação a uma distribuição teórica, mostrando se os dados seguem uma distribuição normal. O resultado é exibido com a função `plt.show()` após a configuração do gráfico com `fig, ax = plt.subplots()`.

fig, ax = plt.subplots()
stats.probplot(dados2, fit=True,   plot=ax)
plt.show()
```

- A imagem acima é um gráfico de probabilidade, que é uma técnica gráfica para avaliar se um conjunto de dados segue ou não uma determinada distribuição, como a distribuição normal. 
- O gráfico tem uma linha vermelha e pontos azuis. 
- O eixo x está rotulado como "Quantis teóricos" e o eixo y está rotulado como "Valores ordenados". 
- A linha vermelha é uma linha reta, que é a linha esperada se os dados seguirem uma distribuição normal. 
- Os pontos azuis representam os pontos de dados, que estão espalhados em torno da linha vermelha. 
- Os pontos de dados desviam da linha vermelha nas caudas, o que sugere que os dados podem não seguir uma distribuição normal.

```python
# Esse código em Python usa a função "shapiro" do módulo "stats" para realizar o teste de normalidade de uma variável chamada "dados2". O teste de Shapiro-Wilk é utilizado para verificar se os dados seguem uma distribuição normal. O resultado da função indicará se os dados são significativamente diferentes de uma distribuição normal (se o valor-p for menor que um nível de significância escolhido) ou se não há evidências para rejeitar a hipótese de normalidade dos dados (se o valor-p for maior que o nível de significância).

stats.shapiro(dados2)
```

O código apresenta os resultados de um teste de normalidade chamado teste de Shapiro-Wilk. O valor "statistic" (estatística) indica a medida de quão próximo os dados estão da distribuição normal, sendo 0.9545 neste caso. O valor "pvalue" (valor p) é extremamente baixo (4.75e-17), sugerindo que os dados não seguem uma distribuição normal, pois o p-valor é menor que o nível de significância típico de 0.05. Portanto, os dados não passam no teste de normalidade de Shapiro-Wilk.
