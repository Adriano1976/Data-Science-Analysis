## Importando as Bibliotecas Necessárias

O código abaixo é usado para criar e visualizar grafos em Python. Primeiro, ele importa as bibliotecas necessárias, igraph e plot, que são usadas para trabalhar com grafos. Com essas bibliotecas, você pode criar e desenhar grafos de forma simples e eficiente em seu programa.

```python
from igraph import Graph
from igraph import plot
```

No código abaixo, estamos criando um grafo direcionado com pesos para representar relações entre quatro pessoas: Fernando, Pedro, José e Antônio. As arestas do grafo representam conexões entre essas pessoas, onde algumas são amigas e outras são inimigas. Cada pessoa tem um peso associado, indicando sua importância. As amizades têm um peso também. Isso nos permite visualizar e analisar as relações e a importância de cada pessoa no contexto das amizades e inimizades estabelecidas.

## Criando um Grafo Direcionado com Pesos de Relações

```python
defineEdges = [(0, 1), (2, 3), (0, 2), (0, 3)]
grafo5 = Graph(edges=defineEdges, directed=True)

labelName = ['Fernando', 'Pedro', 'Jose', 'Antonio'] 
pesos = [40, 30, 30, 25]
grafo5.vs['label'] = labelName
grafo5.vs['peso'] = pesos

friendshipTypes = ['Amigo', 'Inimigo', 'Inimigo', 'Amigo']
arestasPesos = [1, 2, 1, 3]
grafo5.es['TipoAmizade'] = friendshipTypes
grafo5.es['weight'] = arestasPesos
```

## Exibindo Informações Específicas de um Grafo

No código a seguir, utiliza um loop (um tipo de repetição) para percorrer cada "edgesView" (uma parte de um gráfico) no "grafo5" e imprime (mostra na tela) o conteúdo de cada "edgesView". Em resumo, ele está exibindo informações específicas contidas em um gráfico chamado "grafo5" para que você possa vê-las.

```python
for edgesView in grafo5.vs:
    print(edgesView)
```

## Exibindo Informações Sobre as Arestas de um Grafo

 Esse trecho de código abaixo, é usado para mostrar informações sobre as arestas de um grafo. Ele percorre todas as arestas no grafo (representado por 'grafo5') e imprime cada uma delas. As arestas são basicamente conexões entre pontos em um gráfico ou rede, e este código ajuda a visualizá-las.

```python
for edgesView in grafo5.es:
    print(edgesView)
```

## Criando Gráfico Colorido dos Vértices

No código a seguir, estamos colorindo os vértices de um grafo usando a biblioteca Python chamada igraph. Os vértices são representados por cores, e aqui definimos que os quatro vértices do grafo terão cores diferentes: azul, vermelho, amarelo e verde. Em seguida, o código cria um gráfico com essas cores para os vértices e o exibe em uma caixa de 400x400 pixels.

```python
grafo5.vs['cor'] = ['blue', 'red', 'yellow', 'green']
representativeGraph = plot(
    grafo5, 
    bbox = (0, 0, 400, 400),
    vertex_color = grafo5.vs['cor']
)

representativeGraph
```

## Exibindo as Conexões Entre Elementos e Seus Pesos

No código a seguir, cria um gráfico que representa as conexões entre elementos, onde cada conexão tem um peso (espessura da linha) e cada elemento tem uma cor. O gráfico é exibido em uma caixa de 400x400 pixels. Basicamente, ele visualiza as relações entre elementos usando cores e espessuras de linhas para mostrar o quanto essas relações são importantes. É útil para análises visuais de dados.

```python
representativeGraph = plot(
    grafo5,
    bbox=(400, 400),
    edge_width=grafo5.es['weight'],
    vertex_color=grafo5.vs['cor']
)

representativeGraph
```

No código a seguir, estamos criando um gráfico que representa visualmente informações sobre prisões. O gráfico utiliza dados de peso nas arestas (conexões entre elementos) para determinar a espessura das linhas e a cor dos vértices (pontos). Isso nos ajuda a visualizar de forma clara e simples como os elementos estão interligados em relação ao tema das prisões. O resultado final é exibido na variável "representativeGraph".

```python
representativeGraph = plot(
    grafo5, 
    bbox = (400, 400),  
    edge_width = grafo5.es['weight'], 
    vertex_color = grafo5.vs['cor']
)

representativeGraph
```

No código a seguir, estamos criando um gráfico representando uma rede, onde cada ponto (vértice) é dimensionado de acordo com seu peso, as conexões (arestas) têm larguras proporcionais aos pesos e os vértices são coloridos de acordo com uma característica chamada 'cor'. O resultado visual desse código é um gráfico que ajuda a visualizar relações entre elementos, tornando os pontos maiores e as linhas mais largas para destacar informações importantes na rede.

```python
representativeGraph = plot(
     grafo5, 
     bbox = (400, 400), 
     vertex_size = grafo5.vs['peso'],
     edge_width = grafo5.es['weight'],
     vertex_color = grafo5.vs['cor']
)

representativeGraph
```

No código a seguir, cria um gráfico visual que representa dados de forma simples, curvado e atraente. Ele utiliza um conjunto de informações, como o tamanho dos pontos, a espessura das linhas e as cores para destacar características específicas dos dados. O resultado final é um gráfico que facilita a compreensão das relações entre os elementos representados.

```python
representativeGraph = plot(
     grafo5, 
     bbox = (400, 400), 
     vertex_size = grafo5.vs['peso'],
     edge_width = grafo5.es['weight'],
     vertex_color = grafo5.vs['cor'],
     edge_curved = 0.4
)

representativeGraph
```

No código a seguir, estamos criando um gráfico que representa uma rede (grafo). As diferentes partes desse gráfico, como os pontos (vértices) e as linhas que os conectam (arestas), têm características específicas. Os pontos têm tamanhos diferentes com base em seus pesos, cores diferentes e têm a forma de triângulos. As linhas entre eles têm larguras diferentes, determinadas pelo atributo de peso das arestas, e são ligeiramente curvas para uma aparência estilizada. O resultado desse código é a exibição desse gráfico visualmente, permitindo a análise das relações entre os elementos da rede.

```python
representativeGraph = plot(
     grafo5, 
     bbox = (0, 0, 400, 400), 
     vertex_size = grafo5.vs['peso'],
     edge_width = grafo5.es['weight'],
     vertex_color = grafo5.vs['cor'],
     edge_curved = 0.4, 
     vertex_shape = 'triangle'
)

representativeGraph
```
