## Importando as Bibliotecas Necessárias

O código abaixo é usado para criar e visualizar grafos em Python. Primeiro, ele importa as bibliotecas necessárias, igraph e plot, que são usadas para trabalhar com grafos. Com essas bibliotecas, você pode criar e desenhar grafos de forma simples e eficiente em seu programa.

```python
# Importação das bibliotecas
from igraph import Graph
from igraph import plot
```

## Criando um Gráfico Direcionado

Nesse código abaixo, estamos criando um grafo direcionado, que é uma representação de conexões entre pontos, onde as setas indicam a direção das conexões. Em seguida, atribuímos rótulos (ou etiquetas) a cada ponto do grafo. Por fim, o código imprime o grafo na tela, exibindo as conexões e os rótulos dos pontos. Isso pode ser útil para visualizar e entender relações complexas entre elementos.

```python
# Criar um grafo direcionado
arestas = [(0, 1), (1, 2), (2, 3), (3, 0)]
grafo1 = Graph.TupleList(arestas, directed=True)

# Atribuir rótulos aos vértices
grafo1.vs['label'] = list(range(grafo1.vcount()))

# Imprimir o grafo
print(grafo1)
```

## Criando um Gráfico não Direcionado

Nesse código abaixo, estamos criando um gráfico não direcionado, onde os pontos (0, 1), (1, 2), (2, 3) e (3, 0) representam conexões entre diferentes elementos. Em seguida, atribuímos rótulos aos vértices para identificá-los. Por fim, exibimos o gráfico na tela. Isso pode ser útil para visualizar conexões entre elementos em um formato gráfico simples.

```python
# Criando um grafo não direcionado
arestas = [(0, 1), (1, 2), (2, 3), (3, 0)]
grafo2 = Graph(edges=arestas, directed=False)

# Definindo rótulos para os vértices
grafo2.vs['label'] = range(grafo2.vcount())

# Exibindo o grafo
print(grafo2)
```

## Exibindo um Gráfico

Esse código é usado para exibir um gráfico. A função "plot" é usada para desenhar um gráfico chamado "grafo2" em uma área retangular de tamanho 400x400 unidades. É como se você estivesse mostrando um desenho em um quadro, mas aqui é feito em um programa de computador.

```python
# Exibição do gráfico
plot(grafo2, bbox=(0, 0, 400, 400))
```

## Criando um Diagrama Visual não Direcionado

Nesse código a seguir, estamos criando um diagrama visual que representa conexões entre objetos. Primeiro, definimos um gráfico não direcionado (ou seja, as conexões não têm uma direção específica). Em seguida, adicionamos 10 pontos (vértices) ao gráfico e criamos conexões entre alguns deles. Cada ponto é numerado de 0 a 9. Depois, atribuímos números como rótulos aos pontos. Por fim, exibimos o gráfico e o plotamos em um formato visual, mostrando como esses pontos estão conectados.

```python
# Criar um grafo não direcionado
grafo3 = Graph(directed=False)

# Adicionar vértices ao grafo
numero_de_vertices = 10
grafo3.add_vertices(numero_de_vertices)

# Adicionar arestas ao grafo
arestas = [(0, 1), (2, 2), (2, 3), (3, 0)]
grafo3.add_edges(arestas)

# Atribuir rótulos aos vértices
grafo3.vs['label'] = range(numero_de_vertices)

# Imprimir o grafo
print(grafo3)

# Plotar o grafo
plot(grafo3, bbox=(0, 0, 400, 400))
```

## Criando um Gráfico com 7 Pontos Identificados

Nesse código abaixo, estamos criando um gráfico que representa conexões entre sete pontos identificados pelas letras A, B, C, D, E, F e G. Essas conexões são definidas pelas arestas entre esses pontos. Além disso, adicionamos dois pontos adicionais, identificados como "5" e "6", ao gráfico. Cada ponto tem um rótulo que corresponde às letras mencionadas anteriormente. Finalmente, o código cria um gráfico visual para representar essas conexões e pontos em um plano bidimensional.

```python
# Criação do quarto grafo, configurando rótulos personalizados para os vértices
grafo4 = Graph(directed=False)
num_vertices = 7
vertices = ["A", "B", "C", "D", "E", "F", "G"]
arestas = [(0, 1), (1, 2), (2, 3), (3, 4), (4, 0), (0, 2), (2, 1)]

grafo4.add_vertices(num_vertices)
grafo4.add_edges(arestas)

# Adicionando os vértices "5" e "6"
grafo4.add_vertices(["5", "6"])

# Configurando rótulos personalizados para os vértices
grafo4.vs["label"] = vertices

# Plot do grafo
plot(grafo4, bbox=(0, 0, 400, 400))
```
