### Importação das bibliotecas mais importantes

Esse código a seguir, cria e visualiza um gráfico usando a biblioteca igraph. Ele importa as ferramentas necessárias, cria um objeto de gráfico e, em seguida, exibe esse gráfico. O igraph é uma ferramenta poderosa para representar e analisar dados de rede, tornando a visualização de relacionamentos entre elementos mais fácil para análise.

```python

from igraph import Graph
from igraph import plot
```

### Criação de um gráfico direcionado com pesos entre as arestas

Nesse código, estamos criando um gráfico que representa conexões entre diferentes pontos (chamados de nós) usando setas para mostrar a direção das conexões. Além disso, atribuímos pesos a cada conexão para indicar sua importância. Por exemplo, a seta de 'A' para 'C' tem um peso de 2. Essa representação pode ser usada para visualizar redes de várias coisas, como rotas de transporte ou relações entre elementos, e os pesos indicam a força ou custo das conexões.

```python

grafo = Graph(edges=[(0, 2), (0, 1), (1, 4), (1, 5), (2, 3), (6, 7), (3, 7), (4, 7), (5, 6)], directed=True)
grafo.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
grafo.es['weight'] = [2, 1, 2, 1, 2, 1, 3, 1]
```

### Visualização dos resultados

Nesse código, estamos criando uma visualização de dados usando um gráfico chamado "grafo". O gráfico é exibido em uma área retangular de 500x500 pixels. Além disso, estamos adicionando rótulos nas arestas do gráfico que representam os pesos das conexões. Isso é útil para mostrar informações em um formato gráfico compreensível.

```python

plot(
    grafo,
    bbox=(0, 0, 500, 500),
    edge_label=grafo.es['weight']
)
```

### Encontrar o menor caminho entre A - H (retorna os vértices)

Este código a seguir, calcula o menor caminho entre os vértices A e H em um grafo e retorna a sequência de vértices que compõem esse caminho. O resultado será uma lista de vértices que representam a rota mais curta entre os pontos A e H no grafo.

```python

vertice_path = grafo.get_shortest_paths(0, 7, output = 'vpath')

print(vertice_path)
```

### Encontra o caminho mais curto entre os vértices 0 e 7 no grafo

Esse código utiliza um grafo para encontrar o caminho mais curto entre os vértices 0 e 7, representando conexões. Ele retorna as arestas que compõem esse caminho. É como se você estivesse buscando a rota mais curta em um mapa, e o código mostra as estradas que você deve seguir para chegar do ponto 0 ao ponto 7.

```python

edge_path = grafo.get_shortest_paths(0, 7, output='epath')

print(edge_path)
```
