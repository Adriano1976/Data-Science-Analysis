### Importação das bibliotecas mais importantes

Esse código a seguir, cria e visualiza um gráfico usando a biblioteca igraph. Ele importa as ferramentas necessárias, cria um objeto de gráfico e, em seguida, exibe esse gráfico. O igraph é uma ferramenta poderosa para representar e analisar dados de rede, tornando a visualização de relacionamentos entre elementos mais fácil para análise.

```python
from igraph import Graph
from igraph import plot
import igraph
```

### Carrega o grafo a partir de um arquivo no formato GraphML

No código a seguir, faz duas coisas simples: primeiro, ele carrega um gráfico (uma representação visual de conexões) a partir de um arquivo em um formato específico chamado GraphML. Em seguida, ele imprime esse gráfico na tela. Em resumo, o código lê e exibe um gráfico armazenado em um arquivo.

```python
file_path = '../dados/Grafo.graphml'
grafo = igraph.load(file_path)

print(grafo)
```

###  Plota um grafo em um gráfico com tamanho especificado.

No código a seguir, estamos definindo as dimensões de um gráfico com 800 pixels de largura e 800 pixels de altura. Em seguida, usamos essas dimensões para criar e exibir um gráfico chamado 'plotar_graph'. Basicamente, estamos configurando o tamanho do gráfico e depois o exibindo.

```python
width = 800
length = 800
plotar_graph = plot(grafo, bbox = (0, 0, width, length))

plotar_graph
```

### Obtendo as métricas de grau do grafo do total, da entrada e da saída

Esse código a seguir, calcula métricas de grau para um grafo, que é uma representação de conexões entre elementos. O "total_degree" é a quantidade de conexões total de um elemento no grafo, o "grau_entry" é a quantidade de conexões que entram em um elemento, e o "exit_degree" é a quantidade de conexões que saem de um elemento. Isso é útil para entender como os elementos estão interconectados em um grafo.

```python
total_degree = grafo.degree(mode='all')
print(total_degree)
```

```python
grau_entry = grafo.degree(mode='in')
print(grau_entry)
```

```python
exit_degree = grafo.degree(mode='out')
print(exit_degree)
```

### Gerando um gráfico com tamanho dos vértices proporcional ao grau

Nesse código, estamos criando um gráfico onde o tamanho dos pontos (vértices) é determinado pelo número de conexões que cada ponto tem. Quanto mais conexões, maior o ponto. O gráfico é exibido em uma área retangular com largura e altura definidas como 800 unidades. Em seguida, o gráfico é plotado na tela. Essa representação visual ajuda a entender a estrutura das relações entre os pontos em um grafo de maneira simples e intuitiva.

```python
width = 800
length = 800
plotar_graph = plot(grafo, vertex_size = total_degree ,bbox = (0, 0, width, length))

plotar_graph
```

### Encontrar o diâmetro do grafo (a maior distância entre os vértices)

Neste código, estamos determinando o “diâmetro” de um grafo direcionado, que é a distância máxima entre dois pontos no grafo. Em seguida, mostramos o resultado na tela. É uma medida significativa na teoria dos grafos para compreender a extensão máxima das conexões no grafo, o que pode ser útil em análises de redes.

```python
graph_diameter = grafo.diameter(directed = True)
print("Diameter: ", graph_diameter)
```

### Retorno dos vértices que possuem a maior distância entre os pontos do grafo

Neste código, estamos realizando o cálculo e a exibição dos vértices em um grafo que possuem a maior distância entre si. Isso é feito utilizando a função `get_diameter()` do objeto `grafo`, que retorna essa distância. Em seguida, imprimimos a mensagem “Distâncias” seguida do valor calculado. Isso é útil para identificar os pontos mais distantes em um grafo.

```python
greater_distance = grafo.get_diameter()
print("Distâncias: ", greater_distance)
```

### Obter os vizinhos de cada vértice no grafo

Esse código realiza uma ação em um grafo, que é uma representação de conexões entre elementos, onde ele busca e retorna os vizinhos de cada ponto no grafo, ou seja, identifica quais elementos estão diretamente ligados a cada elemento específico no gráfico, possibilitando uma análise das relações entre eles.

```python
get_neighbor = grafo.neighborhood()
get_neighbor
```

### Verificar se o grafo é isomórfico

No código a seguir, estamos verificando se dois grafos (que são basicamente representações de conexões entre pontos) são semelhantes. Para fazer isso, copiamos o grafo original para 'check_graph' e, em seguida, usamos a função 'isomorphic' para comparar os dois grafos e ver se eles têm a mesma estrutura. O resultado, que é impresso na tela, nos dirá se os grafos são semelhantes (isomórficos) ou não, ou seja, se têm a mesma forma básica.

```python
check_graph = grafo
is_isomorph = grafo.isomorphic(check_graph)

print("Os grafos são isomórficos? ",is_isomorph)
```
