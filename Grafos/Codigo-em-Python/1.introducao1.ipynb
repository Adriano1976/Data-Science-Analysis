## Importando as Bibliotecas Necessárias

O código abaixo é usado para criar e visualizar grafos em Python. Primeiro, ele importa as bibliotecas necessárias, igraph e plot, que são usadas para trabalhar com grafos. Com essas bibliotecas, você pode criar e desenhar grafos de forma simples e eficiente em seu programa.

```python
# Importação das bibliotecas
from igraph import Graph
from igraph import plot
```

## Criando um Grafo Direcionado

Nesse código a seguir, estamos criando um grafo direcionado (um conjunto de conexões entre pontos) representado por "grafo1". Ele possui quatro vértices numerados de 0 a 3, e cada vértice é conectado ao próximo na sequência, formando um ciclo. Esses números também são usados como rótulos para os vértices. Finalmente, o código imprime o grafo resultante na tela. É uma maneira básica de visualizar uma estrutura de conexões entre pontos em um programa de computador.

```python
# Definição do grafo com as arestas
grafo1 = Graph(edges = [(0,1),(1,2),(2,3),(3,0)], directed = True)

# Definição do rótulo de cada vértice
grafo1.vs['label'] = range(grafo1.vcount())
print(grafo1)
```

## Criando Gráfico em Plot

Este código a seguir, é usado para criar um gráfico visual a partir de um conjunto de dados chamado "grafo1". A linha "# Gráfico" é um comentário que ajuda a entender o que o código faz. O comando "plot" é responsável por desenhar o gráfico, e a parte "bbox = (0,0,300,300)" define o tamanho e a posição do gráfico na tela. Em resumo, este código gera um gráfico com base nos dados do "grafo1" e o exibe em uma área de 300x300 pixels.

```python
# Gráfico Visual
plot(grafo1, bbox = (0,0,300,300))
```

## Criando um Grafo Direcionado com 8 Nós

Nesse código a seguir, estamos criando um grafo direcionado com oito nós, representados pelos números de 0 a 3, e conexões entre esses nós conforme as tuplas no início do código. Em seguida, adicionamos rótulos aos nós para identificá-los. Por fim, exibimos o grafo em um gráfico bidimensional de 400x400 pixels. Esse código visualiza como esses nós estão interconectados e suas direções.

```python
# Criação do segundo grafo direcionado
arestas = [(0, 1), (1, 2), (2, 3), (3, 0), (0, 3), (3, 2), (2, 1), (1, 0)]
grafo2 = Graph(edges=arestas, directed=True)
grafo2.vs['label'] = range(grafo2.vcount())

plot(grafo2, bbox=(0, 0, 400, 400))
```

## Criando um Gráfico Com 4 Vértices

Nesse código, estamos criando um gráfico que representa conexões entre pontos, chamados de vértices. Primeiro, damos um nome ao gráfico e o configuramos como direcionado, o que significa que as conexões têm uma direção. Em seguida, adicionamos 4 vértices identificados como A, B, C e D. Depois, estabelecemos as conexões entre esses vértices, indicando quais estão conectados. Por fim, exibimos o gráfico em uma caixa com tamanho específico. O gráfico resultante mostrará como esses pontos estão interligados.

```python
# Crie o grafo com um nome descritivo
grafo3 = Graph(directed=True)

# Adicione os vértices com rótulos descritivos
grafo3.add_vertices(4)
grafo3.vs['label'] = ['A', 'B', 'C', 'D']

# Adicione as arestas de forma clara e organizada
grafo3.add_edges([(0, 1), (1, 2), (2, 3), (3, 0), (1, 1)])

# Plote o grafo com um tamanho de caixa especificado
plot(grafo3, bbox=(0, 0, 400, 400))
```

## Criando um Gráfico Direcionado com Ponto Isolado

Este  abaixo cria um gráfico direcionado com conexões entre diferentes pontos. Cada conexão é representada por uma seta que vai de um ponto a outro. Em seguida, adiciona um ponto isolado com o rótulo "5" a esse gráfico. Cada ponto (ou vértice) é rotulado com um número sequencial. Finalmente, o código gera um gráfico visual desse modelo em uma caixa de tamanho 400x400, mostrando como os pontos estão conectados entre si.

```python
# Criar um grafo direcionado com arestas e vértices
arestas = [(0, 1), (1, 2), (2, 3), (3, 0), (1, 1)]
grafo4 = Graph(edges=arestas, directed=True)

# Adicionar um vértice isolado com rótulo "5"
grafo4.add_vertex("5")

# Atribuir rótulos aos vértices
grafo4.vs['label'] = range(grafo4.vcount())

# Plotar o grafo em uma caixa de tamanho 400x400
plot(grafo4, bbox=(0, 0, 400, 400))
```
