### ***Criando uma funo para limpar e converter os valores de uma coluna para numrico***

No cdigo a seguir, est sendo definida a **funo** `clean_numeric_column(df, object)` em Python. Ela recebe dois **parmetros**: `df` (um **objeto** DataFrame do **mdulo** `pandas`) e `object` (uma string com o nome da coluna). Dentro da funo,  verificado com `pd.api.types.is_numeric_dtype()` se a coluna j  numrica; se no for, os caracteres no numricos so removidos usando `str.replace()` e a coluna  convertida para valores numricos com `pd.to_numeric()`. Caso j seja numrica,  exibida uma mensagem informativa. A linha `pd.options.mode.chained_assignment = None` previne avisos de atribuio encadeada no pandas.

```python
def clean_numeric_column(df, object):

    if not pd.api.types.is_numeric_dtype(df[object]):
        df[object] = pd.to_numeric(
            df[object].str.replace(r'\D', '', regex=True)
        )
    else:
        print(f"Column '{object}' is already numeric.")

pd.options.mode.chained_assignment = None
```

No cdigo a seguir, est sendo chamada a **funo** `clean_numeric_column()` com o **objeto** `base_dropna` (DataFrame do **mdulo** `pandas`) e a coluna escolhida como **parmetro**. A funo garante que todos os valores da coluna sejam numricos, removendo caracteres no numricos e convertendo para tipo numrico, se necessrio. Em seguida, a **funo** `unique()`  aplicada sobre `base_dropna['coluna-escolhida']` para retornar um **objeto** `numpy.ndarray` com todos os valores distintos da coluna j limpos e convertidos.


```python
clean_numeric_column(base_dropna, 'Idade')
base_dropna['Idade'].unique()
```

```python
clean_numeric_column(base_dropna, 'Pessoas_Familia')
base_dropna["Pessoas_Familia"].unique()
```

A lista `value_conditions` define os valores atribudos a cada condio, e o **parmetro** `default` indica o valor padro. Por fim, o **mtodo** de indexao `[['colunas']]` exibe apenas as trs colunas escolhidas.

```python
conditions = [
    (base_dropna['Violencia_Parceiro'] == 'Yes') & (base_dropna['score_violencia'] >= 1),
    (base_dropna['Violencia_Parceiro'] == 'Yes') & (base_dropna['score_violencia'] < 1),
    (base_dropna['Violencia_Parceiro'] == 'No') & (base_dropna['score_violencia'] < 1),
    (base_dropna['Violencia_Parceiro'] == 'No') & (base_dropna['score_violencia'] >= 1)
]

value_conditions = ['Yes', 'No', 'Yes', 'No']
base_dropna['consciente_violencia'] = np.select(conditions, value_conditions, default='Sem valor')
base_dropna[['Violencia_Parceiro', 'score_violencia', 'consciente_violencia']]
```

### ***Verificando se cada coluna escolhida segue uma distribuio normal usando a funo `kstest`***

No cdigo a seguir, est sendo aplicado o **teste estatstico** Kolmogorov-Smirnov atravs da **funo** `kstest`, que pertence ao **mdulo** `scipy.stats`, para verificar se cada coluna escolhida (`Idade`, `Pessoas_Familia` e `score_violencia`) segue uma distribuio normal. Para isso, em cada iterao do **loop for**, so calculados o valor da estatstica (`stat`) e o valor-p (`p`), usando como **parmetros** a mdia e o desvio padro da prpria coluna. Em seguida, a estrutura condicional `if/else` avalia a hiptese nula (H0), imprimindo se a coluna se ajusta ou no  normalidade.

```python
for column in base_dropna[['Idade', 'Pessoas_Familia', 'score_violencia']]:
    stat, p = kstest(base_dropna[column], 'norm', args=(base_dropna[column].mean(), base_dropna[column].std()))
    print(f'\nkstest Test for {column}: Statistics={stat:.3f}, p-value={p:.3f}')
    
    if p > 0.05:
        print(f'{column} parece ser normalmente distribuda (falha em rejeitar H0)')
    else:
        print(f'{column} no parece ser normalmente distribuda (rejeita H0)')
```

### ***Visualizando as correlaes significativas entre colunas usando a funo `spearmanr`***

No cdigo a seguir, est sendo criada a lista `columns_numeric` como objeto contendo as variveis `Idade`, `Pessoas_Familia` e `score_violencia`, e em seguida so realizados laos `for` (estrutura de repetio) para percorrer pares de colunas, aplicando a funo `spearmanr` da biblioteca **scipy.stats** (mdulo) que retorna a correlao e o p-valor; os parmetros `col1` e `col2` recebem cada par de variveis, enquanto o mtodo `print` exibe os resultados, avaliando se a correlao  estatisticamente significativa conforme a condio `p_value < 0.05`.

```python
columns_numeric = ['Idade', 'Pessoas_Familia', 'score_violencia']
print("\n Significant Correlation (|r| > 0.3 e p-value < 0.05)")

for i in range(len(columns_numeric)):
    for j in range(i + 1, len(columns_numeric)):
        col1 = columns_numeric[i]
        col2 = columns_numeric[j]
        corr, p_value = spearmanr(base_dropna[col1], base_dropna[col2])
        print(f'\nSpearman correlation between {col1} and {col2}: corr: {corr:.3f}, p-value: {p_value:.3f}')
        
        if p_value < 0.05:
            print(f'There is a significant correlation between {col1} and {col2} (reject H0)')
        else:
            print(f'There is no significant correlation between {col1} and {col2} (fail to reject H0)')
```

### ***Agrupando os valores numricos para decidir a significncia estatstica com a funo `f_oneway`***

No cdigo a seguir, est sendo utilizado o mtodo `groupby` do objeto **DataFrame** `base_dropna` para agrupar os valores da coluna `Idade` de acordo com a varivel categrica `Religiao`, formando uma lista de sries (`base_groups`) sem valores nulos por meio da funo `dropna`. Em seguida, a funo `f_oneway` do mdulo **scipy.stats**  chamada como funo de teste ANOVA, recebendo os grupos como parmetros com o operador `*`. Os objetos `f_stat` e `p_val` armazenam respectivamente o valor da estatstica F e o p-valor, que so avaliados em um teste de hiptese para decidir sobre a significncia estatstica.

```python
base_groups = [group['Idade'].dropna() for name, group in base_dropna.groupby('Religiao', observed=False)]
f_stat, p_val = f_oneway(*base_groups)
print(f'\n ANOVA Test: F-statistic: {f_stat:.3f}, p-value: {p_val:.3f}')

if p_val < 0.05:
    print('H diferena significativa entre as mdias dos grupos (rejeita H0)')
else:
    print('No h diferena significativa entre as mdias dos grupos (falha em rejeitar H0)')
```

No cdigo a seguir, est sendo utilizada a funo `pd.concat` do mdulo **pandas** para concatenar todos os grupos de idades armazenados na lista de objetos `base_groups`, criada anteriormente a partir do mtodo `groupby` do **DataFrame**. Em seguida,  aplicado o mtodo `mean`, que  uma funo para clculo da mdia, retornando a idade mdia geral. O resultado  armazenado no objeto `general_median_age` e exibido na tela com o uso da funo `print`, utilizando uma f-string para formatar a sada com trs casas decimais.

```python
general_median_age = pd.concat(base_groups).mean()
print(f'\n General Mean Age: {general_median_age:.3f}')
```

No cdigo a seguir, est sendo calculada a soma total dos quadrados (SST), medida usada em ANOVA para avaliar a variabilidade total dos dados. O mdulo **pandas** fornece a funo `pd.concat` para juntar os grupos de idades em um nico objeto, do qual  subtrado o valor armazenado no objeto `general_median_age`. O resultado dessa diferena  elevado ao quadrado com o operador `** 2`, e em seguida, a funo `sum` agrega todos os valores, retornando a variabilidade total. O resultado final  exibido com a funo `print`, formatado em trs casas decimais.


```python
ss_total = sum((pd.concat(base_groups) - general_median_age) ** 2)
print(f'Total Sum of Squares (SST): {ss_total:.3f}')
```

O trecho de cdigo calcula a **soma dos quadrados entre os grupos (SSB)**, que mede a variabilidade explicada pelas diferenas entre as mdias de cada grupo em relao  mdia geral. Para cada grupo em `base_groups`, a diferena entre a mdia do grupo e a mdia geral  elevada ao quadrado, multiplicada pelo tamanho do grupo (`len(group)`), e depois somada para todos os grupos. O resultado final  exibido formatado em trs casas decimais.


```python
ss_between = sum([len(group) * (group.mean() - general_median_age) ** 2 for group in base_groups])
print(f'\n Sum of Squares Between Groups: {ss_between:.3f}')
```

Esse cdigo calcula o **eta squared ()**, uma medida de tamanho de efeito da ANOVA. Ele  obtido dividindo a soma dos quadrados entre grupos (`ss_between`) pela soma total dos quadrados (`ss_total`). O resultado indica a proporo da varincia total que pode ser explicada pelas diferenas entre as mdias dos grupos, variando de 0 a 1.


```python
eta_squared = ss_between / ss_total
print(f'\n Eta Squared: {eta_squared:.3f}')
```

### ***Agrupando os valores numricos para decidir a significncia estatstica com a funo `kruskal`***

Em seguida, a **funo** `kruskal` do **mdulo** `scipy.stats`  chamada, recebendo os grupos como **parmetro** com o operador `*`, retornando a estatstica H e o p-valor, armazenados nos objetos `stat` e `p_val`, que so exibidos com a funo `print` para avaliar a diferena significativa entre os grupos.

### ***Calculando o tamanho de efeito `Eta` para o teste de Kruskal-Wallis***

No cdigo a seguir, est sendo calculado o **tamanho de efeito** Eta para o teste de Kruskal-Wallis. O objeto `K` recebe o nmero de grupos em `base_groups` com a funo `len`, enquanto `N` soma o total de observaes usando uma list comprehension e `len(group)` para cada grupo. A frmula `(f_stat - K + 1) / (N - K)` calcula o Eta, armazenando o resultado no objeto `eta2_kruskal`, que  exibido com a funo `print`, formatado com trs casas decimais.


```python
K = len(base_groups)
N = sum([len(group) for group in base_groups])
eta2_kruskal = (f_stat - K + 1) / (N - K)
print(f'\n Eta (Kruskal-Wallis): {eta2_kruskal:.3f}')
```

### ***Realizando o teste qui-quadrado de independncia entre as variveis usando a funo `chi2_contingency`***

No cdigo a seguir, est sendo realizado o **teste qui-quadrado de independncia** usando a funo `chi2_contingency` do mdulo **scipy.stats**, aplicada ao objeto `table_crosstab`, que  um **DataFrame** do mdulo **pandas**. A funo retorna quatro objetos: `chi2` (estatstica do teste), `p` (valor-p), `dof` (graus de liberdade) e `expected` (frequncias esperadas). Em seguida, a funo `print` exibe os resultados formatados, e uma estrutura condicional `if/else` avalia se existe associao significativa entre as variveis, considerando o critrio `p < 0.05`.


```python
chi2, p, dof, expected = chi2_contingency(table_crosstab)
print(f'Chi-Squared Test: Chi2={chi2:.3f}, p-value={p:.3f}')

if p < 0.05:
    print('H associao significativa entre as variveis')
else:
    print('No h associao significativa entre as variveis')
```

### ***Avaliando a fora da associao entre as variveis numricas usando a funo criada `cramers_v`***

No cdigo a seguir, est sendo definida a **funo** `cramers_v`, que recebe como **parmetro** uma tabela de contingncia (`confusion_matrix`) do **mdulo** pandas. Dentro da funo, a **funo** `chi2_contingency` do **mdulo** `scipy.stats` calcula a estatstica qui-quadrado, `n` recebe o total de observaes somando todos os elementos, e `k` obtm o menor nmero de categorias menos um. O retorno  o clculo de Cramr's V usando `np.sqrt`. O objeto `v` armazena o resultado, exibido com `print`, e uma estrutura condicional avalia a fora da associao entre as variveis.


```python
def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    k = min(confusion_matrix.shape) - 1
    return np.sqrt(chi2 / (n * k))

v = cramers_v(table_crosstab)
print(f"Cramr's V: {v:.3f}")

if p >= 0.5:
    print('H associao forte entre as variveis')    
elif p >= 0.3:
    print('H associao moderada entre as variveis')
else:
    print('Ha  associao forte entre as variveis')
```
