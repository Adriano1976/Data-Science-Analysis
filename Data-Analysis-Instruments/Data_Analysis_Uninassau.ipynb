### ***Criando uma função para limpar e converter os valores de uma coluna para numérico***

No código a seguir, está sendo definida a **função** `clean_numeric_column(df, object)` em Python. Ela recebe dois **parâmetros**: `df` (um **objeto** DataFrame do **módulo** `pandas`) e `object` (uma string com o nome da coluna). Dentro da função, é verificado com `pd.api.types.is_numeric_dtype()` se a coluna já é numérica; se não for, os caracteres não numéricos são removidos usando `str.replace()` e a coluna é convertida para valores numéricos com `pd.to_numeric()`. Caso já seja numérica, é exibida uma mensagem informativa. A linha `pd.options.mode.chained_assignment = None` previne avisos de atribuição encadeada no pandas.

```python
def clean_numeric_column(df, object):

    if not pd.api.types.is_numeric_dtype(df[object]):
        df[object] = pd.to_numeric(
            df[object].str.replace(r'\D', '', regex=True)
        )
    else:
        print(f"Column '{object}' is already numeric.")

pd.options.mode.chained_assignment = None
```

No código a seguir, está sendo chamada a **função** `clean_numeric_column()` com o **objeto** `base_dropna` (DataFrame do **módulo** `pandas`) e a coluna escolhida como **parâmetro**. A função garante que todos os valores da coluna sejam numéricos, removendo caracteres não numéricos e convertendo para tipo numérico, se necessário. Em seguida, a **função** `unique()` é aplicada sobre `base_dropna['coluna-escolhida']` para retornar um **objeto** `numpy.ndarray` com todos os valores distintos da coluna já limpos e convertidos.


```python
clean_numeric_column(base_dropna, 'Idade')
base_dropna['Idade'].unique()
```

```python
clean_numeric_column(base_dropna, 'Pessoas_Familia')
base_dropna["Pessoas_Familia"].unique()
```

### ***Criando tabelas de contingências que cruzam 2 variáveis ou mais usando a função `crosstab`***

No código a seguir, está sendo utilizada a função `pd.crosstab` do módulo **pandas** para criar uma tabela de contingência que cruza a variável **Sexo** com a variável **Estado\_Civil**. A classe implícita é `DataFrame`, já que o resultado é retornado nesse formato. Os parâmetros `margins=True` e `margins_name='Total'` adicionam uma linha e coluna extras chamadas “Total”, que somam os valores por linha e coluna. Assim, o objeto `sex_civil` passa a armazenar a contagem de combinações entre sexo e estado civil, permitindo análises descritivas de frequência.


```python
sex_civil = pd.crosstab(base_dropna['Sexo'], base_dropna['Estado_Civil'], margins=True, margins_name='Total')
sex_civil
```

```python
sex_civil_perc = pd.crosstab(base_dropna['Sexo'],
                             base_dropna['Estado_Civil'],
                             margins=True,
                             margins_name='Total',
                             normalize='all').round(4) * 100
sex_civil_perc
```

No código a seguir, está sendo utilizada a função **`pd.crosstab`** do módulo **pandas** para gerar duas tabelas: a primeira (`sex_civil`) com contagens absolutas e a segunda (`sex_civil_perc`) com percentuais normalizados. O parâmetro **`margins=True`** adiciona totais e **`margins_name='Total'`** define o nome da linha/coluna de soma. O método **`.round(1)`** arredonda os percentuais e a multiplicação por 100 converte para porcentagem. Em seguida, a operação de concatenação (`astype(str) + ...`) combina os valores absolutos e relativos em uma única tabela (`combined`), que é exibida com o objeto **`display`**.


```python
sex_civil = pd.crosstab(base_dropna['Sexo'],
                        base_dropna['Estado_Civil'],
                        margins=True,
                        margins_name='Total')

sex_civil_perc = pd.crosstab(base_dropna['Sexo'],
                             base_dropna['Estado_Civil'],
                             margins=True,
                             margins_name='Total',
                             normalize='all').round(1) * 100

combined = sex_civil.astype(str) + ' (' + sex_civil_perc.astype(str) + '%)'
display(combined)
```

A lista `value_conditions` define os valores atribuídos a cada condição, e o **parâmetro** `default` indica o valor padrão. Por fim, o **método** de indexação `[['colunas']]` exibe apenas as três colunas escolhidas.

```python
conditions = [
    (base_dropna['Violencia_Parceiro'] == 'Yes') & (base_dropna['score_violencia'] >= 1),
    (base_dropna['Violencia_Parceiro'] == 'Yes') & (base_dropna['score_violencia'] < 1),
    (base_dropna['Violencia_Parceiro'] == 'No') & (base_dropna['score_violencia'] < 1),
    (base_dropna['Violencia_Parceiro'] == 'No') & (base_dropna['score_violencia'] >= 1)
]

value_conditions = ['Yes', 'No', 'Yes', 'No']
base_dropna['consciente_violencia'] = np.select(conditions, value_conditions, default='Sem valor')
base_dropna[['Violencia_Parceiro', 'score_violencia', 'consciente_violencia']]
```

### ***Verificando se cada coluna escolhida segue uma distribuição normal usando a função `kstest`***

No código a seguir, está sendo aplicado o **teste estatístico** Kolmogorov-Smirnov através da **função** `kstest`, que pertence ao **módulo** `scipy.stats`, para verificar se cada coluna escolhida (`Idade`, `Pessoas_Familia` e `score_violencia`) segue uma distribuição normal. Para isso, em cada iteração do **loop for**, são calculados o valor da estatística (`stat`) e o valor-p (`p`), usando como **parâmetros** a média e o desvio padrão da própria coluna. Em seguida, a estrutura condicional `if/else` avalia a hipótese nula (H0), imprimindo se a coluna se ajusta ou não à normalidade.

```python
for column in base_dropna[['Idade', 'Pessoas_Familia', 'score_violencia']]:
    stat, p = kstest(base_dropna[column], 'norm', args=(base_dropna[column].mean(), base_dropna[column].std()))
    print(f'\nkstest Test for {column}: Statistics={stat:.3f}, p-value={p:.3f}')
    
    if p > 0.05:
        print(f'{column} parece ser normalmente distribuída (falha em rejeitar H0)')
    else:
        print(f'{column} não parece ser normalmente distribuída (rejeita H0)')
```

### ***Visualizando as correlações significativas entre colunas usando a função `spearmanr`***

No código a seguir, está sendo criada a lista `columns_numeric` como objeto contendo as variáveis `Idade`, `Pessoas_Familia` e `score_violencia`, e em seguida são realizados laços `for` (estrutura de repetição) para percorrer pares de colunas, aplicando a função `spearmanr` da biblioteca **scipy.stats** (módulo) que retorna a correlação e o p-valor; os parâmetros `col1` e `col2` recebem cada par de variáveis, enquanto o método `print` exibe os resultados, avaliando se a correlação é estatisticamente significativa conforme a condição `p_value < 0.05`.

```python
columns_numeric = ['Idade', 'Pessoas_Familia', 'score_violencia']
print("\n Significant Correlation (|r| > 0.3 e p-value < 0.05)")

for i in range(len(columns_numeric)):
    for j in range(i + 1, len(columns_numeric)):
        col1 = columns_numeric[i]
        col2 = columns_numeric[j]
        corr, p_value = spearmanr(base_dropna[col1], base_dropna[col2])
        print(f'\nSpearman correlation between {col1} and {col2}: corr: {corr:.3f}, p-value: {p_value:.3f}')
        
        if p_value < 0.05:
            print(f'There is a significant correlation between {col1} and {col2} (reject H0)')
        else:
            print(f'There is no significant correlation between {col1} and {col2} (fail to reject H0)')
```

### ***Agrupando os valores numéricos para decidir a significância estatística com a função `f_oneway`***

No código a seguir, está sendo utilizado o método `groupby` do objeto **DataFrame** `base_dropna` para agrupar os valores da coluna `Idade` de acordo com a variável categórica `Religiao`, formando uma lista de séries (`base_groups`) sem valores nulos por meio da função `dropna`. Em seguida, a função `f_oneway` do módulo **scipy.stats** é chamada como função de teste ANOVA, recebendo os grupos como parâmetros com o operador `*`. Os objetos `f_stat` e `p_val` armazenam respectivamente o valor da estatística F e o p-valor, que são avaliados em um teste de hipótese para decidir sobre a significância estatística.

```python
base_groups = [group['Idade'].dropna() for name, group in base_dropna.groupby('Religiao', observed=False)]
f_stat, p_val = f_oneway(*base_groups)
print(f'\n ANOVA Test: F-statistic: {f_stat:.3f}, p-value: {p_val:.3f}')

if p_val < 0.05:
    print('Há diferença significativa entre as médias dos grupos (rejeita H0)')
else:
    print('Não há diferença significativa entre as médias dos grupos (falha em rejeitar H0)')
```

No código a seguir, está sendo utilizada a função `pd.concat` do módulo **pandas** para concatenar todos os grupos de idades armazenados na lista de objetos `base_groups`, criada anteriormente a partir do método `groupby` do **DataFrame**. Em seguida, é aplicado o método `mean`, que é uma função para cálculo da média, retornando a idade média geral. O resultado é armazenado no objeto `general_median_age` e exibido na tela com o uso da função `print`, utilizando uma f-string para formatar a saída com três casas decimais.

```python
general_median_age = pd.concat(base_groups).mean()
print(f'\n General Mean Age: {general_median_age:.3f}')
```

No código a seguir, está sendo calculada a soma total dos quadrados (SST), medida usada em ANOVA para avaliar a variabilidade total dos dados. O módulo **pandas** fornece a função `pd.concat` para juntar os grupos de idades em um único objeto, do qual é subtraído o valor armazenado no objeto `general_median_age`. O resultado dessa diferença é elevado ao quadrado com o operador `** 2`, e em seguida, a função `sum` agrega todos os valores, retornando a variabilidade total. O resultado final é exibido com a função `print`, formatado em três casas decimais.


```python
ss_total = sum((pd.concat(base_groups) - general_median_age) ** 2)
print(f'Total Sum of Squares (SST): {ss_total:.3f}')
```

O trecho de código calcula a **soma dos quadrados entre os grupos (SSB)**, que mede a variabilidade explicada pelas diferenças entre as médias de cada grupo em relação à média geral. Para cada grupo em `base_groups`, a diferença entre a média do grupo e a média geral é elevada ao quadrado, multiplicada pelo tamanho do grupo (`len(group)`), e depois somada para todos os grupos. O resultado final é exibido formatado em três casas decimais.


```python
ss_between = sum([len(group) * (group.mean() - general_median_age) ** 2 for group in base_groups])
print(f'\n Sum of Squares Between Groups: {ss_between:.3f}')
```

Esse código calcula o **eta squared (η²)**, uma medida de tamanho de efeito da ANOVA. Ele é obtido dividindo a soma dos quadrados entre grupos (`ss_between`) pela soma total dos quadrados (`ss_total`). O resultado indica a proporção da variância total que pode ser explicada pelas diferenças entre as médias dos grupos, variando de 0 a 1.


```python
eta_squared = ss_between / ss_total
print(f'\n Eta Squared: {eta_squared:.3f}')
```

### ***Agrupando os valores numéricos para decidir a significância estatística com a função `kruskal`***

No código a seguir, está sendo criado o objeto `base_groups` como uma lista de séries, usando o método `groupby` do **DataFrame** `base_dropna` (módulo **pandas**) para separar os valores da coluna **score\_violencia** conforme a coluna **consciente\_violencia**, removendo valores nulos com o método `dropna()`. Em seguida, a **função** `kruskal` do **módulo** `scipy.stats` é chamada, recebendo os grupos como **parâmetro** com o operador `*`, retornando a estatística H e o p-valor, armazenados nos objetos `stat` e `p_val`, que são exibidos com a função `print` para avaliar a diferença significativa entre os grupos.


```python
base_groups = [group['score_violencia'].dropna() for name, group in base_dropna.groupby('consciente_violencia')]
stat, p_val = kruskal(*base_groups)
print(f'Kruskal-Wallis Test: H-statistic: {stat:.3f}, p-value: {p_val:.3f}')
```

### ***Calculando o tamanho de efeito `Eta²` para o teste de Kruskal-Wallis***

No código a seguir, está sendo calculado o **tamanho de efeito** Eta² para o teste de Kruskal-Wallis. O objeto `K` recebe o número de grupos em `base_groups` com a função `len`, enquanto `N` soma o total de observações usando uma list comprehension e `len(group)` para cada grupo. A fórmula `(f_stat - K + 1) / (N - K)` calcula o Eta², armazenando o resultado no objeto `eta2_kruskal`, que é exibido com a função `print`, formatado com três casas decimais.


```python
K = len(base_groups)
N = sum([len(group) for group in base_groups])
eta2_kruskal = (f_stat - K + 1) / (N - K)
print(f'\n Eta² (Kruskal-Wallis): {eta2_kruskal:.3f}')
```

### ***Realizando o teste qui-quadrado de independência entre as variáveis usando a função `chi2_contingency`***

No código a seguir, está sendo realizado o **teste qui-quadrado de independência** usando a função `chi2_contingency` do módulo **scipy.stats**, aplicada ao objeto `table_crosstab`, que é um **DataFrame** do módulo **pandas**. A função retorna quatro objetos: `chi2` (estatística do teste), `p` (valor-p), `dof` (graus de liberdade) e `expected` (frequências esperadas). Em seguida, a função `print` exibe os resultados formatados, e uma estrutura condicional `if/else` avalia se existe associação significativa entre as variáveis, considerando o critério `p < 0.05`.


```python
chi2, p, dof, expected = chi2_contingency(table_crosstab)
print(f'Chi-Squared Test: Chi2={chi2:.3f}, p-value={p:.3f}')

if p < 0.05:
    print('Há associação significativa entre as variáveis')
else:
    print('Não há associação significativa entre as variáveis')
```

### ***Avaliando a força da associação entre as variáveis numéricas usando a função criada `cramers_v`***

No código a seguir, está sendo definida a **função** `cramers_v`, que recebe como **parâmetro** uma tabela de contingência (`confusion_matrix`) do **módulo** pandas. Dentro da função, a **função** `chi2_contingency` do **módulo** `scipy.stats` calcula a estatística qui-quadrado, `n` recebe o total de observações somando todos os elementos, e `k` obtém o menor número de categorias menos um. O retorno é o cálculo de Cramér's V usando `np.sqrt`. O objeto `v` armazena o resultado, exibido com `print`, e uma estrutura condicional avalia a força da associação entre as variáveis.


```python
def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    k = min(confusion_matrix.shape) - 1
    return np.sqrt(chi2 / (n * k))

v = cramers_v(table_crosstab)
print(f"Cramér's V: {v:.3f}")

if p >= 0.5:
    print('Há associação forte entre as variáveis')    
elif p >= 0.3:
    print('Há associação moderada entre as variáveis')
else:
    print('Ha á associação forte entre as variáveis')
```
